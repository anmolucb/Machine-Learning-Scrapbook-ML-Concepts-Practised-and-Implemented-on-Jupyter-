```python
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
```

```python
data=pd.read_csv('diabetes_data.csv')
data_pca=data.drop(columns='Diabetes_012')
```

```python
from sklearn.preprocessing import StandardScaler
#to standardise all values using std dev=1 and mean=0
```

```python
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data_pca)
```

```python
from sklearn.decomposition import PCA
#PCA is a feature reduction technique and reduces number of features. 
#It is useful in reducing computation load as we are dealing with smaller dataset
```

```python
#reducing it to just n dimensions
pca=PCA(n_components=2)
```

```python
data_scaled=pca.fit_transform(data_scaled)
```

```python
data_scaled
```

```python
#displaying how the 2 dimensional data is able to classify between Diabeters and No Diabetes
plt.scatter(data_scaled[:,0],data_scaled[:,1],c=data['Diabetes_012'])
```

If you want an exact amount of variance to be kept in data after applying PCA, specify a float between 0 and 1 to the hyperparameter n_components.

This is the easiest way to select the best number of principal components for the dataset. For example, if you want to keep 75% of the variance in the original data after applying PCA, you can specify the float 0.75 to the hyperparameter n_components as follows.

from sklearn.decomposition import PCA
pca = PCA(n_components=0.75)

Then, the algorithm automatically selects the best number of principal components that keep 85% of the variance in the original data. If you want to know how many components that the algorithm has selected, run the following line of code.

```python
pca=PCA(n_components=0.85)
data_scaled=pca.fit_transform(data_scaled)
data_scaled
```

```python
pca.n_components_
```
